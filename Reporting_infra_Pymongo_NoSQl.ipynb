```python
import pymongo
import csv
import json
import pandas as pd
from pymongo import MongoClient
client = MongoClient()
client = MongoClient('localhost', 27017)
```

```python
db=client.parcelled
db.collection_names()
```

**Example of Updating a record**


```python
db.bookings.update_many({"status":"Inititated"},{"$set":{"status":"Init"}})
```

```python
cursor=db.bookings.aggregate([{"$group" : {"_id" : "$status", "num_status" : {"$sum" : 1}}}])
df =  pd.DataFrame(list(cursor))
df
```

```python
##Taking data out from inside an object and choosing only specific columns 
cursor1=db.bookings.find({"consignee.city":"Indore"},{'booking_id':1,'_id':0,'booking_items.is_cod':1})
```

```python
cursor1
df1 =  pd.DataFrame(list(cursor1))
df1
```

```python
###Working with datetime data in Mongodb. Also using greater than or less than operators. 


from datetime import datetime 
start = datetime(2015, 3, 24, 7, 51,4)
end = datetime(2015, 4, 24, 7, 52,4)

cursor1=db.bookings.find({"consignor_pickup.city":"Bangalore","confirmed_at":{"$gt":start,"$lte":end}},{'booking_id':1,'_id':0,'confirmed_at':1,'booking_items.is_cod':1,'booking_items.is_fragile':1,'booking_items.name':1,"status":1})
df1 =  pd.DataFrame(list(cursor1))
df1
```

```python
type(df1['booking_items'][0])
```

```python
def remove_list(data):
    new_dict = {}
    for item in data:
        if 'is_cod' in item:
            new_dict['is_cod'] = item['is_cod']
        if 'name' in item:
            new_dict['name'] = item['name']
        if 'is_fragile' in item:
            new_dict['is_fragile'] = item['is_fragile']
    return new_dict
df1['booking_items']=df1['booking_items'].apply(remove_list)
```

```python
df2=df1['booking_items'].apply(pd.Series)
```

```python
pd.concat([df1.drop(['booking_items'], axis=1),df2], axis=1)
```
